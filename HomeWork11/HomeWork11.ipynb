```python
import math
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
```

```python
fx = -5*x**2 + 10*x - 150
fx
```

```python
# –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫
plot(fx)
```

```python
# –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∫–æ—Ä–Ω–∏
x1, x2 = solve(fx)
round(x1.evalf(),3), round(x2.evalf(),3)
```

```python
# –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –ø—Ä–æ–º–µ–∂—É—Ç–∫–∏, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º f > 0
solve(fx>0)
```

```python
# –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –ø—Ä–æ–º–µ–∂—É—Ç–∫–∏, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º f < 0
solve(fx<0)
```

```python
# –ù–∞–π—Ç–∏ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã, –Ω–∞ –∫–æ—Ç–æ—Ä—ã—Ö —Ñ—É–Ω–∫—Ü–∏—è —É–±—ã–≤–∞–µ—Ç
solve(diff(fx)<0)
```
